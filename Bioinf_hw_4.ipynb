### Белок TLR3
Ключевой компонент врожденного и адаптивного иммунитета. TLR (Toll-подобные рецепторы) контролируют иммунный ответ хозяина против патогенов путем распознавания молекулярных паттернов, характерных для микроорганизмов. TLR3 - нуклеотид-чувствительный TLR, который активируется двухцепочечной РНК - признаком вирусной инфекции. Действует через адаптер TRIF/TICAM1, приводя к активации NF-kappa-B, ядерной транслокации IRF3, секреции цитокинов и воспалительной реакции.

* Homo sapiens,
